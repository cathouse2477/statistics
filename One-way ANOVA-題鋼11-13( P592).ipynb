```python
import numpy as np 
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

df=pd.DataFrame({
    'operator': np.tile(np.repeat( ["operator-1","operator-2","operator-3","operator-4","operator-5"],1),3),
     'machine': np.repeat( ["machine-1","machiner-2","machine-3"],5),
     'output' : [53,47,46,50,49,51,51,49,54,50,61,55,52,58,54]})
df
#df[:10]
## 'sun': np.tile(np.repeat(['low', 'med', 'high'], 5), 2),
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

#perform two-way ANOVA
model = ols('output ~ C(operator) + C(machine) ', data=df).fit()
sm.stats.anova_lm(model, typ=2)
```
